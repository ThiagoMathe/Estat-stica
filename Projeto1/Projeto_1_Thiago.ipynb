- (B) Qual a proporção de pessoas ilesas e de feridos graves por mês em cada um dos anos considerados?

```python
print("\nProporção de pessoas ilesas e de feridos do ano de 2018:")
dataframe_2018['data_inversa'] = pd.to_datetime(dataframe_2018['data_inversa'])
ilesos2018 = dataframe_2018.groupby(dataframe_2018['data_inversa'].dt.month)['ilesos'].sum()
graves2018 = dataframe_2018.groupby(dataframe_2018['data_inversa'].dt.month)['feridos_graves'].sum()
pessoas2018 = dataframe_2018.groupby(dataframe_2018['data_inversa'].dt.month)['pessoas'].sum()
conta_ilesos2018 = ilesos2018/pessoas2018
conta_graves2018 = graves2018/pessoas2018
conta_Total2018 = (ilesos2018+graves2018)/pessoas2018
df_conta2018 = pd.concat([conta_ilesos2018,conta_graves2018,conta_Total2018],axis=1)
df_conta2018.columns=["Total ilesos","Total graves","Total graves e ilesos"]
df_conta2018.index = (["Janeiro","fervereiro","Março","Abril","Maio","junho","julho",
                  "Agosto","Setembro","Outubro","Novembro","Dezembro"])
df_conta2018
```

```python
print("\nProporção de pessoas ilesas e de feridos do ano de 2019:")
dataframe_2019['data_inversa'] = pd.to_datetime(dataframe_2019['data_inversa'])
ilesos2019 = dataframe_2019.groupby(dataframe_2019['data_inversa'].dt.month)['ilesos'].sum()
graves2019 = dataframe_2019.groupby(dataframe_2019['data_inversa'].dt.month)['feridos_graves'].sum()
pessoas2019 = dataframe_2019.groupby(dataframe_2019['data_inversa'].dt.month)['pessoas'].sum()
conta_ilesos2019 = ilesos2019/pessoas2019
conta_graves2019 = graves2019/pessoas2019
conta_Total2019 = (ilesos2019+graves2019)/pessoas2019
df_conta2019 = pd.concat([conta_ilesos2019,conta_graves2019,conta_Total2019],axis=1)
df_conta2019.columns=["Total ilesos","Total graves","Total graves e ilesos"]
df_conta2019.index = (["Janeiro","fervereiro","Março","Abril","Maio","junho","julho",
                  "Agosto","Setembro","Outubro","Novembro","Dezembro"])
df_conta2019
```

```python
print("\nProporção de pessoas ilesas e de feridos do ano de 2020:")
dataframe_2020['data_inversa'] = pd.to_datetime(dataframe_2020['data_inversa'])
ilesos2020 = dataframe_2020.groupby(dataframe_2020['data_inversa'].dt.month)['ilesos'].sum()
graves2020 = dataframe_2020.groupby(dataframe_2020['data_inversa'].dt.month)['feridos_graves'].sum()
pessoas2020 = dataframe_2020.groupby(dataframe_2020['data_inversa'].dt.month)['pessoas'].sum()
conta_ilesos2020 = ilesos2020/pessoas2020
conta_graves2020 = graves2020/pessoas2020
conta_Total2020 = (ilesos2020+graves2020)/pessoas2020
df_conta2020 = pd.concat([conta_ilesos2020,conta_graves2020,conta_Total2020],axis=1)
df_conta2020.columns=["Total ilesos","Total graves","Total graves e ilesos"]
df_conta2020.index = (["Janeiro","fervereiro","Março","Abril","Maio","junho","julho",
                  "Agosto","Setembro","Outubro","Novembro","Dezembro"])
df_conta2020
```

Usando o comando groupby para agrupar grandes quantidades de dados,ja o dt.month usei para extrair o mês do dataframe e a função .sum para somar todos os casos. Juntei os dados por mês de cada caso fiz a contas dos ilesos e graves pelo número de pessoas envolvidas separadamente e fiz novamento porém com ilesos e graves juntos. fiz as junções das variáveis e defini as colunas com seus nomes e as linhas em ordem pelos meses.
